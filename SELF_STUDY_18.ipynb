```python
with open("istiklal.txt", "r", encoding="utf-8") as file:
  lines = file.readlines()

counter = 0

with open("istiklal.txt", "w", encoding="utf-8") as file:
  for i in lines:
    counter +=1
    if counter % 4 == 0:
      file.write(i+"\n")
    else:
      file.write(i)

with open("istiklal.txt", "r", encoding="utf-8") as filem:
  print(filem.read())
```
