```python
withdrawal = "1"
deposit = "2"
balance_inquiry = "3"
exit = "4"
balance = 1000
print("Pls choose an option:")

print("Press 1 for withdrawal.")
print("Press 2 for deposit cash.")
print("Press 3 for balance enquiry.")
print("Press 4 for exit.")

while True:
  option = input("Pls make a choice.")
  if option == deposit:
    balance += int(input("Pls enter the amount: "))

  elif option == balance_inquiry:
    print(f"Your current balance is £{balance:}.")
  elif option==withdrawal:
    print(f"Your current balance is {balance}")
    minus = int(input("Pls enter the amount: "))
    if minus > balance:
      minus = int(input("Enter lower amount: "))
    balance -= minus
    print(f"Your current balance is {balance}")
  elif option==exit:
    break
```

Try to find out if a number you get from the user is perfect. A number is called a "perfect number" if the sum of its divisors is equal to itself. For example, 6 is a perfect number. (1 + 2 + 3 = 6)

```python
number = int(input("Pls enter your number: "))

divisors=0

for i in range(1,number):
  if number%i==0:
    divisors +=i
if divisors==number:
  print(f"{number} is a perfect number.")
```

A number from the user will be set to be "Armstrong. For example, if a number has 4 digits and is equal to the 4th (3rd power for 3-digit numbers) of each of its surroundings, this number is called the "Armstrong" number. etc.(1634 = 1^4 + 6^4 + 3^4 + 4^4)

```python
number = input("Pls enter your number:")
print(number)
power = len(number)
check = 0
print(power)
for i in number:
  check += int(i)**power
if check==int(number):
  print(f"{number} is an Armstrong number.") 
else:
  print(f"{number} is not an Armstrong number.")
```

In each while loop, take a number from the user and add the numbers entered by the user to a variable named "sum". When the user presses the "q" key, terminate the loop and print the "sum variable" to the screen.

```python
sum = 0

while True:
  a = input("Pls enter a number, and q to exit:")
  if a!="q":
    sum +=int(a)
  else:
      break
print(f"Sum of the numbers you entered is {sum}")
```
